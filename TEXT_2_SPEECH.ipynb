```python
from gtts import gTTS
from IPython.display import Audio
from IPython.display import display
import pandas as pd
import time
import librosa
import audioread
```

```python
L1 = []
with open('KOMPAS1.txt') as dat:
  dat = dat.read()
  for x in dat.split('\n'):
    L1.append(x)


```

```python
for x in L1:
  if len(x) < 2:
    L1.remove(x)
```

```python
L1
```

```python
DATA = pd.DataFrame()
```

```python
DATA['TEXT'] = L1
DATA['SGN'] = 0
```

```python
DATA = DATA.dropna()
```

```python
def simpanBaca(XXX):
  try:
    speech = gTTS(text= str(XXX), lang='id', slow=False)
    speech.save('baca.wav')
    wn = Audio('baca.wav', autoplay=True)

    display(wn)
    display(str(XXX))
    with audioread.audio_open('baca.wav') as f:
      print(f.duration)
      time.sleep(f.duration)
  except:
    pass  
```

```python
for x in range(len(DATA)):
  A1 = DATA['TEXT'][x]
  simpanBaca(A1)
  
```
